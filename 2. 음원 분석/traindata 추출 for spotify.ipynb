```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
cid = 'bebe6e09d6be4631a30f1c81a8d5e8e1'
secret = '06aed97340eb46bea139aac73f9f55bb'
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
 
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

def get_features(song): # get track id information 
    track_info = sp.search(q=song, type='track', market='JP') 
    track_id = track_info["tracks"]["items"][0]["id"] # track id로 정보를 빼온다.
    # get audio_feature 
    features = sp.audio_features(tracks=[track_id]) 
    acousticness = features[0]["acousticness"] 
    danceability = features[0]["danceability"] 
    energy = features[0]["energy"]
    instrumental = features[0]["instrumentalness"]
    key = features[0]["key"]
    loudness = features[0]["loudness"]
    valence = features[0]["valence"] 
    mode = features[0]["mode"]
    tempo = features[0]["tempo"]
    timetempo = features[0]["time_signature"]
    
    result = {"acousticness" : acousticness,
              "danceability" : danceability, # 춤추기 적합한 정도
              "energy" : energy,             # 화려하고 노이즈가 많을수록 1에 가까움(0~1사이값)
              "loudness" : loudness,         # 소리의 화려함을 (데시벨 값으로)
              "key"      : key,             # 0 =C, 1 = C#
              "valence" : valence,         # 음원의 밝을 정도를 나타내는 값 밝고 행복하면 값이 높고 슬프면 낮음
              "mode" : mode,                 # major = 1, minor = 0
              "tempo" : tempo,               # bpm 으로 표시한 전체예상 템포
              "time_signature" : timetempo}  # 박자표(마디당 혹은 소절에 비트 수)       
    return result

```

### blue_eyed feat

```python
blue_songn.columns = ['artist_name','song_name']
blue_songn.drop('artist_name', axis=1, inplace=True)
blue_songn
```

```python
feature_list=[]
for i in tq(range(0,4488)):
    try:
        get = get_features(blue_songn['song_name'][i])
        feature_list.append(get)
    except:
        feature_list.append({'acousticness': 0,
                         'danceability': 0,
                         'energy': 0,
                         'loudness': 0,
                         'key': 0,
                         'valence': 0,
                         'mode': 0,
                         'tempo': 0,
                         'time_signature': 0})
```
